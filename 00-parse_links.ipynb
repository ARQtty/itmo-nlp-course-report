```python
import re
```

```python
link_pattern = re.compile(r'https:\/\/[\w\.\d\/-]+')
text = open('./raw/links/raw_links_all.txt').read()
```

```python
links = re.findall(link_pattern, text)
len(links)
```

```python
sites = ('moika78', 'spbdnevnik', 'fontanka')

for site in sites:
    with open(f'./raw/links/links_{site}.txt', 'w') as f:
        site_links = list(filter(lambda s: site in s, links))
        f.write('\n'.join(site_links))
        print(f'{site} {len(site_links)} links')
```
